Importamos la libreria matplotlib para graficar.
Adicionalmente se vincula o importa la clase para crear y calcular las neuronas.
En esta clase se genera las caminatas aleatorias con probabilidades establecidas y que posteriormente analizan
el estado de cada una de ellas vinculadas entre si.

```python
from random_walk import WalkRandom
import matplotlib.pyplot as plt
```

A continuacion, creamos las tres neuronas con la clase WalkRandom creada, estableciendo los parámetros con las cuales se harán los cálculos necesarios para obtener las caminatas para cada una de ellas.
Se establece número de iteraciones, g, L, C, p y pf.

```python
clase = WalkRandom(neuronas=3, iteraciones=500, C=1, g=0.01, L=30, p=0.5, pf=0.9)
```

Mediante la función procesa() de la clase WalkRandom, calcula las caminatas de las neuronas que son vinculadas entre sí mediante el valor de g.

```python
clase.procesa()
```

Mediante la función obtener_caminata() obtenemos las posiciones de cada una de las neuronas.
La función obtiene un parámetro que corresponde al número de neurona empezando desde cero.

```python
walk0 = clase.obtener_caminata(0)
walk1 = clase.obtener_caminata(1)
walk2 = clase.obtener_caminata(2)
```

Mediante el uso de las funciones que matplotlib nos brinda podemos realizar el grafico de cada una de las neuronas por separado y adicionalmente las 3 solapadas en un solo grafico.

```python
plt.plot(walk0, label = "n0", color="green")
plt.show()
```

```python
plt.plot(walk1, label = "n1", color="red")
plt.show()
```

```python
plt.plot(walk2, label = "n2", color="blue")
plt.show()
```

```python
plt.plot(walk0, label = "n0", color="green")
plt.plot(walk1, label = "n1", color="red")
plt.plot(walk2, label = "n2", color="blue")
plt.show()
```
